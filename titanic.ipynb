```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket'],axis=1,inplace=True)
```

```python
df.isna().sum()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Cabin']=df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Pclass']=df['Pclass'].fillna(df['Pclass'].mean())
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
pc=df['Pclass'].value_counts()
plt.bar(pc.index,pc.values)
```

```python
sx=df['Sex'].value_counts()
plt.bar(sx.index,sx.values)
```

```python
dfe=pd.get_dummies(df[['Sex','Cabin','Embarked']],drop_first=True,dtype=int)
dfe
```

```python
df2=pd.concat([df,dfe],axis=1)
df2
```

```python
df2.drop(['Sex','Cabin','Embarked'],axis=1,inplace=True)
```

```python
df2.columns
```

```python
df2
```

```python
x=df2.drop(['Survived'],axis=1)
```

```python
y=df2['Survived']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(x_train)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy Score is ",accuracy_score(y_test,y_pred))
```

```python
print('''confusion matrix''', confusion_matrix(y_test,y_pred))
```
